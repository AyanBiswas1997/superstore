```python
#ordinal_encoding
Item_Fat_Content_map={"Low Fat":0,"Regular":1,"LF":0,"low fat":0,"reg":1} 
Item_Type_map={"Dairy":0,"Soft Drinks":1,"Meat":2,"Fruits and Vegetables":3,"Household":4,"Baking Goods":5,"Snack Foods":6,"Frozen Foods":7,"Breakfast":8,"Health and Hygiene":9,"Hard Drinks":10,"Canned":11,"Breads":12,"Starchy Foods":13,"Others":14,"Seafood":15}
Outlet_Size={"Small":0,"Medium":1,"High":2}
Outlet_Location_Type={"Tier 1":0,"Tier 2":1,"Tier 3":2} 
Outlet_Type={"Supermarket Type1":0,"Supermarket Type2":1,"Supermarket Type3":2,"Grocery Store":3}
```

```python
data["Item_Fat_Content"]=data["Item_Fat_Content"].map(Item_Fat_Content_map)
data["Item_Type"]=data["Item_Type"].map(Item_Type_map)
data["Outlet_Size"]=data["Outlet_Size"].map(Outlet_Size)
data["Outlet_Location_Type"]=data["Outlet_Location_Type"].map(Outlet_Location_Type)
data["Outlet_Type"]=data["Outlet_Type"].map(Outlet_Type)
```
